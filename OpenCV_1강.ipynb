## 영상의 표현 방법   
영상의 해상도는 가로 x 세로로 표현하며 행렬에서는 세로 x 가로의 형태로 표현한다.   
* 그레이스케일 영상
    - 밝기 정보만으로 구성된 영상
    - 밝기 정보는 0 ~ 255의 값으로 256 단계를 표현
    - Python에서 데이터타입 numpy.uint8을 이용하여 표현
* 트루컬러 영상
    - 색상 정보를 가지는 영상으로 일반적으로 컬러 영상이라하면 이것을 지칭함
    - 색의 정보는 Red, Green, Blue로 구성되며 밝기 정보의 값과 동일하며 256^3 가지 색상을 표현할 수 있음
    - Python에서 3차원 배열의 형태를 이용하여 표현

## 영상 파일 형식의 특징
* BMP
    - 픽셀 데이터를 압축하지 않아 용량이 큼
    - 구조가 단순해 별도 라이브러리 없이 직접 io프로그래밍이 가능
* JPG
    - 픽셀 데이터를 압축한 형태로 화질의 손실을 가짐
    - 압축률이 좋아 디지털 카메라 사진 포멧에 주로 사용
* PNG
    - BMP와 JPG의 특징이 혼합된 형태로 압축이 되면서도 화질의 손실이 없음
    - 투명도 지원
